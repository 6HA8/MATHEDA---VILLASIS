# Highest **Critic_Score**

```python
df_w = df.copy()
```

```python
hgh_cs = df_w.sort_values('critic_score', ascending=False)
hgh_cs
```

```python
df_w['title'] = df_w.groupby('title').cumcount().astype(str).radd(df_w['title'] + '_')
hgh_cs = df_w.sort_values('critic_score', ascending=False).head(10)
hgh_cs
```

```python
plt.figure(figsize=(10, 5))
plt.bar(hgh_cs['title'], hgh_cs['critic_score'], color=['purple', 'gray'])
plt.xlabel('Game Title')
plt.ylabel('Critic Scores')
plt.title('Top 8 Games by Critic Score')
plt.xticks(rotation=90)
plt.show()
```
